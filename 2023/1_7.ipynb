```python
def check_position(pos,a=False):
    if a:
        if ((not pos.isnumeric()) and (not pos == ".")):
            return True 
        else:
            False
    else: 
        if pos == "*":
            return True 
        else:
            False


def check_adj(data, i,j, mode_a = True):
    r = len(data)
    c = len(data[0])
    return_list = []
    if i != 0: 
        if check_position(data[i-1][j]):
            return_list.append((i-1, j))

    if j != 0: 
        if check_position(data[i][j-1]):
            return_list.append((i, j-1))

    if i!= (r-1):
        if check_position(data[i+1][j]):
            return_list.append((i+1, j))

    if j != (c-1):
        if check_position(data[i][j+1]):
            return_list.append((i, j+1))
        
    # Diagonals
    if (i != 0) and (j != 0): 
        if check_position(data[i-1][j-1]):
            return_list.append((i-1, j-1))

    if (i != (r-1)) and (j != 0): 
        if check_position(data[i+1][j-1]):
            return_list.append((i+1, j-1))

    if (i != 0) and (j != (c-1)): 
        if check_position(data[i-1][j+1]):
            return_list.append((i-1, j+1))

    if (i != (r-1)) and (j != (c-1)): 
        if check_position(data[i+1][j+1]):
            return_list.append((i+1, j+1))
    if mode_a:
        if len(return_list) != 0:
            return True
        else:
            return False
    else: 
        return return_list
```

```python
import operator
```

```python
def parse(d, joker_rule=False):
    data = [x.split(" ") for x in d]
    for x in range(len(data)): 
        data[x][0] = list(data[x][0])
        data[x][1] = int(data[x][1])
        for y in range(len(data[x][0])):
            if data[x][0][y] == "T":
                data[x][0][y] = 10
            if data[x][0][y] == "J":
                if joker_rule:
                    data[x][0][y] = -1
                else:
                    data[x][0][y] = 11
            if data[x][0][y] == "Q":
                data[x][0][y] = 12
            if data[x][0][y] == "K":
                data[x][0][y] = 13
            if data[x][0][y] == "A":
                data[x][0][y] = 14
            else:
                data[x][0][y] = int(data[x][0][y])
        data[x][0] = data[x][0]
    return data
```

```python
def check_hand_value(h, joker_rule=False):

    if joker_rule: 
        a,b = np.unique([x for x in h if x != -1], return_counts=True)
        if len(b) == 0: 
            return 7 # 5 jokers
        best = a[b.argmax()]
        h = [x if (x != -1) else best for x in h]

    l = len(set(h))

    if l == 1:
        return 7 
    if l == 2:
        if np.unique(h, return_counts=True)[1].max() == 4:
            return 6
        else:
            return 5
    if l ==3: 
        if np.unique(h, return_counts=True)[1].max() == 3:
            return 4
        else:
            return 3    
    if l == 4: 
        return 2
    else: 
        return 1
```

```python
def compare(d, joker_rule=False):
    ordering = []
    for x in d:
        ordering.append([check_hand_value(x[0], joker_rule=joker_rule)] + x[0] + [x[1]])
    sorted_list = sorted(ordering, key=operator.itemgetter(0, 1,2,3,4,5))
    return (np.array(sorted_list)[:,-1] * np.arange(1,len(sorted_list)+1)).sum()
```

```python
d = parse(data, joker_rule=True)
compare(d, joker_rule=True)
```
