```python
from tools import load
import numpy as np
import re
import math
import matplotlib.pyplot as plt
```

```python
data = load(line_splitter=",", to_list=False)
options = [x.strip() for x in data[0]]
patterns = [x[0].strip() for x in data[2:]]
```

```python
def add_option(p, potential):
    new_potential = []
    for item in potential: 
        remain = p[len(item):]
        for available in [x for x in options if x[:len(x)] == remain[:len(x)]]:
            new_potential.append(item + available)
    return new_potential
```

```python
res = []
for n ,p in enumerate(patterns):
    if n%100 == 0:
        print(n)
    potential = [""]
    while  len(potential) :
        potential = add_option(p, potential)
        potential = set(potential)
        if p in potential: 
            res.append(p)
            break
len(res)
```
